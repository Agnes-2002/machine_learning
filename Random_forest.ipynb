```python
# import libaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading data
df=pd.read_csv("D:\data analytics\data set\credit_train.csv")
print(df.head(20))
```

```python
df.drop(columns=['Loan ID','Customer ID',],inplace=True)
```

```python
df.info()
```

```python
# cleaning data
df.dropna()
print(df.isna().sum())
print(df.duplicated().sum())
print(df.duplicated())
print(df.drop_duplicates(inplace=True))
```

```python
df=df.rename(columns={'Current Loan Amount':'Current_Loan_Amount','Credit Score':'Credit_Score','Annual Income':'Annual_Income','Monthly Debt':'Monthly_Debt','Years of Credit History':'Years_of_Credit_History','Months since last delinquent':'Months_since_last_delinquent','Number of Open Accounts':'Number_of_Open_Accounts','Number of Credit Problems':'Number_of_Credit_Problems','Current Credit Balance':'Current_Credit_Balance','Maximum Open Credit':'Maximum_Open_Credit','Tax Liens':'Tax_Liens'})
                      
                      
```

```python
df.describe()
```

```python
corr=df.corr(numeric_only=True)
print(corr)
sns.heatmap(corr)
plt.show()
```

```python
# finding outliers
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

```python
# Removing outliers
df['z_score'] = (df['Maximum_Open_Credit'] - df['Maximum_Open_Credit'].mean()) / df['Maximum_Open_Credit'].std()
threshold=1
df= df[df['z_score'].abs() <= threshold]
df= df.drop(columns=['z_score'])
print(df)
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

```python
df['z_score'] = (df['Annual_Income'] - df['Annual_Income'].mean()) / df['Annual_Income'].std()
threshold=1
df= df[df['z_score'].abs() <= threshold]
df= df.drop(columns=['z_score'])
print(df)
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

```python
#  Assigning dependent and independent variable
x=df.iloc[:,1]
y=df.iloc[:,0]
x=pd.DataFrame(x)
y=pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
rf = RandomForestClassifier()
rf_model = rf.fit(x_train,y_train)
rf_model
```

```python
y_pred_train = rf_model.predict(x_train)
y_pred_test = rf_model.predict(x_test)
accuracy_score(y_train,y_pred_train),accuracy_score(y_test,y_pred_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(n_estimators=50, learning_rate=1)
model.fit(x_train, y_train)
Y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, Y_pred)
print('Accuracy:', (accuracy*100))
```
